```python
import pandas as pd
import datetime as dt
```

```python
inout = pd.read_csv("DataSets/INOUT.csv")
line = pd.read_csv("DataSets/INOUTLINE.csv", low_memory=False)
prodins = pd.read_csv("DataSets/PRODUCTINSTANCE.csv", encoding='cp1252', low_memory=False)
prod = pd.read_csv("DataSets/PRODUCTS.csv", low_memory=False)
transitem = pd.read_csv("DataSets/TRANSFER_ITEM.csv", low_memory=False)
transitemd = pd.read_csv("DataSets/TRANSFER_ITEM_D.csv", low_memory=False)
```

```python
pattern = "%m/%d/%Y %H:%M"
selectedYear = 2015
count = 0
endYear = False
while endYear != True:
    for i in range(len(inout['CREATED'])):
        if str(inout['CREATED'][i]) == 'nan': continue
        try:
            if(dt.datetime.strptime(inout['CREATED'][i], pattern).year == selectedYear):
                count+=1
        except ValueError:
            continue
    print(count, "\trecords in year ", selectedYear)
    selectedYear += 1
    count = 0
    if selectedYear == 2023: endYear = True
```

```python
dictionary = {}
for i in range(len(line['M_PRODUCT_ID'])):
    if str(line['M_PRODUCT_ID'][i]) == 'nan': continue
    try:
        int(line['M_PRODUCT_ID'][i])
        if(int(line['M_PRODUCT_ID'][i]) in dictionary):
            dictionary[int(line['M_PRODUCT_ID'][i])] += 1
        else:
            dictionary[int(line['M_PRODUCT_ID'][i])] = 1
    except ValueError:
        continue
print(dictionary)
```

```python
countDictionary = {}
valueDictionary = {}
for i in range(len(prodins['ACCT_AC_HOLDING_ID'])):
        if str(prodins['ACCT_AC_HOLDING_ID'][i]) == 'nan': continue
        try:
            if int(prodins['ACCT_AC_HOLDING_ID'][i]) in countDictionary:
                countDictionary[int(prodins['ACCT_AC_HOLDING_ID'][i])] += 1
            else:
                countDictionary[int(prodins['ACCT_AC_HOLDING_ID'][i])] = 1
            if str(prodins['BOOKVALUE'][i]) == 'nan': continue
            if int(prodins['ACCT_AC_HOLDING_ID'][i]) in valueDictionary:
                valueDictionary[int(prodins['ACCT_AC_HOLDING_ID'][i])] += int(prodins['BOOKVALUE'][i])
            else:
                valueDictionary[int(prodins['ACCT_AC_HOLDING_ID'][i])] = int(prodins['BOOKVALUE'][i])
        except ValueError:
            continue
print(countDictionary)
print(valueDictionary)
```
