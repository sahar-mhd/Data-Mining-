Validity

```python
print('index\tINOUT_ID\tISACTIVE')
inOutTrueNumbers = 0
isactiveTrueNumbers = 0
for i in range(max(len(inout['INOUT_ID']), len(inout['ISACTIVE']))):
    try:
        int(inout['INOUT_ID'][i])
        print(i, '\t\tTrue', end='')
        inOutTrueNumbers += 1
    except ValueError:
        print(i, '\t\tFalse', end='')
    if str(inout['ISACTIVE'][i]) == 'Y' or str(inout['ISACTIVE'][i]) == 'N':
        print('\t\tTrue')
        isactiveTrueNumbers += 1
    else:
        print('\t\tFalse')
print("Percent of Validity for INOUT_ID(type checking):",
      (inOutTrueNumbers * 100) / len(inout['INOUT_ID']))
print("Percent of Validity for ISACTIVE:",
      (isactiveTrueNumbers * 100) / len(inout['ISACTIVE']))
```

```python
import datetime as dt
from datetime import datetime
documentdateTrueNumbers = 0
pattern = "%m/%d/%Y %H:%M"
pattern2 = "%m/%d/%Y %H:%M:%S"
print('index\tDOCUMENTDATE')
num = 0
for i in range(len(inout['DOCUMENTDATE'])):
    if str(inout['DOCUMENTDATE'][i]) == 'nan':
        continue
    try:
        date = dt.datetime.strptime(inout['DOCUMENTDATE'][i], pattern)
        if 2022 < date.year < 0 or 1 > date.month > 12 or 1 > date.day > 31 or 24 < date.hour < 0 or 0 > date.minute > 60:  # or date > datetime.today().strftime(
            # pattern):
            num = 2
    except ValueError:
        num += 1
    try:
        dt.datetime.strptime(inout['DOCUMENTDATE'][i], pattern2)
    except ValueError:
        num += 1
    print(i, "\t\t", num == 0 or num == 1)
    if num == 0 or num == 1:
        documentdateTrueNumbers += 1
    num = 0
print("Percent of Validity for DOCUMENTDATE:",documentdateTrueNumbers * 100 / len(inout['DOCUMENTDATE']))
```

```python
ids = inout['INOUT_ID']
print(ids.duplicated(keep=False))
numberOfDuplicatedIds = 0
for i in range(len(inout['INOUT_ID'])):
    if str(ids.duplicated(keep=False)[i]) == 'False':
        numberOfDuplicatedIds+=1
print("Percent of Validity for INOUT_ID(uniqueness): ",numberOfDuplicatedIds * 100 / len(inout['INOUT_ID']))
```

Currentness

```python
print('index\tUPDATED')
numberOfUpdatedRows = 0
for i in range(len(inout['UPDATED'])):
    if str(inout['UPDATED'][i]) == 'nan':
        continue
    try:
        if dt.datetime.strptime(inout['UPDATED'][i], pattern) > dt.datetime.strptime('1/1/2020 00:00', pattern):
            print(i, '\t\t', True)
            numberOfUpdatedRows += 1
        else:
            print(i, '\t\t', False)
    except ValueError:
        continue
print('Number of Updated Rows: ',numberOfUpdatedRows * 100 / len(inout['UPDATED']))
```

Consistency

```python
numberOfComplianceRules = 0
for i in range(min(len(inout['CREATED']), len(inout['UPDATED']), len(inout['DOCUMENTDATE']), len(inout['LETTERDATE']))):
    if str(inout['CREATED'][i]) == 'nan' or str(inout['UPDATED'][i]) == 'nan' or str(
            inout['DOCUMENTDATE'][i]) == 'nan' or str(inout['LETTERDATE'][i]) == 'nan':
        continue
    try:
        print(i, "\t",
              dt.datetime.strptime(inout['CREATED'][i], pattern) <= dt.datetime.strptime(
                  inout['UPDATED'][i], pattern),
              "\t", dt.datetime.strptime(inout['CREATED'][i], pattern) <= dt.datetime.strptime(inout['DOCUMENTDATE'][i],
                                                                                               pattern), "\t",
              dt.datetime.strptime(inout['CREATED'][i], pattern) <= dt.datetime.strptime(inout['LETTERDATE'][i],
                                                                                         pattern))
        if dt.datetime.strptime(inout['CREATED'][i], pattern) <= dt.datetime.strptime(
                  inout['UPDATED'][i], pattern) : numberOfComplianceRules += 1
    except ValueError:
        continue
print("Number of Compliance Rules in Consistency : Created  > Updated: ", numberOfComplianceRules * 100 / len(inout['CREATED']))
```
